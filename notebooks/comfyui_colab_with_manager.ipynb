### Run ComfyUI with cloudflared (Recommended Way)




```python
!wget -P ~ https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i ~/cloudflared-linux-amd64.deb

import subprocess
import threading
import time
import socket
import urllib.request

# Kill any lingering cloudflared processes from previous runs in this session
!pkill -f cloudflared 2>/dev/null || true
time.sleep(1)

def wait_for_url_live(url, timeout=30):
    """Poll the URL until it responds, or give up after timeout seconds."""
    start = time.time()
    while time.time() - start < timeout:
        try:
            req = urllib.request.Request(url, method='HEAD')
            resp = urllib.request.urlopen(req, timeout=5)
            if resp.status < 500:
                return True
        except Exception:
            pass
        time.sleep(2)
    return False

def launch_tunnel(port, max_attempts=3):
    for attempt in range(1, max_attempts + 1):
        print(f"\n[Tunnel attempt {attempt}/{max_attempts}] Launching cloudflared...")
        p = subprocess.Popen(
            ["cloudflared", "tunnel", "--url", f"http://127.0.0.1:{port}"],
            stdout=subprocess.PIPE, stderr=subprocess.PIPE
        )
        found_url = None
        for line in p.stderr:
            l = line.decode()
            if "trycloudflare.com " in l:
                found_url = l[l.find("http"):].strip()
                break

        if not found_url:
            print("No URL printed by cloudflared, retrying...")
            p.terminate()
            time.sleep(2)
            continue

        print(f"Got URL: {found_url}  — verifying it responds...")
        if wait_for_url_live(found_url):
            print(f"\n✅ CONFIRMED WORKING URL: {found_url}\n")
            return found_url, p
        else:
            print(f"URL did not respond within timeout, killing and retrying...")
            p.terminate()
            time.sleep(2)

    print("❌ Failed to establish a working tunnel after all attempts.")
    return None, None

def iframe_thread(port):
    while True:
        time.sleep(0.5)
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        if result == 0:
            sock.close()
            break
        sock.close()
    print("\nComfyUI finished loading, launching cloudflared...\n")
    launch_tunnel(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```

### Run ComfyUI with colab iframe (use only in case the previous way with localtunnel doesn't work)

You should see the ui appear in an iframe. If you get a 403 error, it's your firefox settings or an extension that's messing things up.

If you want to open it in another window use the link.

Note that some UI features like live image previews won't work because the colab iframe blocks websockets.

```python
import threading
import time
import socket
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("to open it in a window you can open this link here:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
